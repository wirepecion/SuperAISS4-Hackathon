```python
count = 0

import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Define the folder path
folder_path = r'raw_data\110'

# List all .npy files in the folder
file_list = [f for f in os.listdir(folder_path) if f.endswith('.npy')]

# Define the sampling frequency (fs)
fs = 250  # Adjust this to your actual sampling frequency

# Plot each file's data
for file_name in file_list:
    file_path = os.path.join(folder_path, file_name)
    # Load the numpy array
    last_data = np.load(file_path)
    last_data = last_data.T
    # Average the signals across channels
    avg_data = np.mean(last_data, axis=0)
    # Create a new figure for each file with a specified size
    plt.figure(figsize=(15, 5))  # Adjust the figure size as needed

    # Plot the combined spectrogram
    plt.subplot(1, 2, 1)
    plt.specgram(avg_data, Fs=fs, NFFT=256, noverlap=128, cmap='viridis')
    plt.title(f'{file_name} - Combined EEG Channels')
    plt.xlabel('Time [s]')
    plt.ylabel('Frequency [Hz]')
    plt.colorbar(label='Intensity [dB]')

    # Plot the frequency graph
    plt.subplot(1, 2, 2)
    freqs, psd = welch(avg_data, fs=fs, nperseg=1024, noverlap=128)
    plt.semilogy(freqs, psd)
    plt.xlabel('Frequency [Hz]')
    plt.ylabel('Power Spectral Density [V**2/Hz]')
    plt.title('Frequency Graph')

    plt.tight_layout()
    plt.show()

    if count == 3:
        break
    count += 1
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
from PIL import Image

# Define the folder path
folder_path = r'split_merge_class\\150'

# List all .npy files in the folder
file_list = [f for f in os.listdir(folder_path) if f.endswith('.npy')]

# Define the sampling frequency (fs)
fs = 250  # Adjust this to your actual sampling frequency

# Create a folder to store the frequency graph images
os.makedirs('rest', exist_ok=True)

# Plot each file's data
for file_name in file_list:
    file_path = os.path.join(folder_path, file_name)
    # Load the numpy array
    last_data = np.load(file_path)
    # Average the signals across channels
    avg_data = np.mean(last_data, axis=0)
    try:
        # Compute the frequency graph
        freqs, psd = welch(avg_data, fs=fs, nperseg=256, noverlap=128)

        # Create a figure for the frequency graph
        plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
        plt.semilogy(freqs, psd)
        plt.xlabel('Frequency [Hz]')
        plt.ylabel('Power Spectral Density [V**2/Hz]')
        plt.title('Frequency Graph')

        # Save the figure as an image
        image_path = os.path.join('rest', f'{file_name}.png')
        plt.savefig(image_path, dpi=100, bbox_inches='tight')
        plt.close()

        # Convert the image to a numpy array
        image = Image.open(image_path)
        image_array = np.array(image)

        # Save the numpy array as a .npy file
        #np.save(os.path.join('frequency_graphs', f'{file_name}.npy'), image_array)

        #print(f'Saved frequency graph for {file_name}')
    except:
        print('Error')
```
