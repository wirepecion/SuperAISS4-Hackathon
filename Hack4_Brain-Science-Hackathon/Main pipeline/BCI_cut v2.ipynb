### Split Ratio and Merge Cleansing

```python
import os
import numpy as np
from scipy.signal import butter, filtfilt, iirnotch

# Make sure the required directories exist
os.makedirs('./split_merge_class_clean/110', exist_ok=True)
os.makedirs('./split_merge_class_clean/120', exist_ok=True)
os.makedirs('./split_merge_class_clean/150', exist_ok=True)

# Define the band-pass filter
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data, axis=0)
    return y

# Define the notch filter
def notch_filter(data, freq, fs, quality=30):
    nyquist = 0.5 * fs
    notch_freq = freq / nyquist
    b, a = iirnotch(notch_freq, quality)
    y = filtfilt(b, a, data, axis=0)
    return y

import os
import numpy as np
from scipy.signal import butter, filtfilt, iirnotch
from sklearn.decomposition import FastICA

# Sampling frequency (you need to set the correct value based on your data)
fs = 250  # example value, replace with your actual sampling frequency

for num in range(0, len(data_time_series)):
    data = np.load('./train/train/' + data_time_series[num])
    time = np.load('./train/train/' + data_time_stamps[num])
    label_time = np.load('./train/train/' + label_time_stamps[num])
    label = np.load('./train/train/' + label_time_series[num])

    # Apply band-pass filter
    data = bandpass_filter(data, 0.1, 100, fs)

    # Apply notch filter
    data = notch_filter(data, 50, fs)

    # Apply ICA to remove artifacts or noise
    ica = FastICA(n_components=8)  # Assuming you have 8 channels/features
    ica_data = ica.fit_transform(data)

    # Calculate the number of subplots needed
    num_subplots = len(label_time) - 1

    for j in range(num_subplots):
        mask = (time >= label_time[j]) & (time < label_time[j+1])
        mask = mask & (time - label_time[j] >= 2) & (time - label_time[j] < 5)
        sub_data = ica_data[mask]

        if label[j, 0] == 110:
            np.save('./split_merge_class_clean/110/' + str(num) + '_' + str(j) + '_110.npy', sub_data)
        elif label[j, 0] == 120:
            np.save('./split_merge_class_clean/120/' + str(num) + '_' + str(j) + '_120.npy', sub_data)
        else:
            np.save('./split_merge_class_clean/150/' + str(num) + '_' + str(j) + '_150.npy', sub_data)

    # Save the last subgraph
    mask = time >= label_time[-1]
    mask = mask & (time - label_time[-1] >= 2) & (time - label_time[-1] < 5)
    sub_data = ica_data[mask]

    if label[-1, 0] == 110:
        np.save('./split_merge_class_clean/110/' + str(num) + '_' + str(num_subplots) + '_110.npy', sub_data)
    elif label[-1, 0] == 120:
        np.save('./split_merge_class_clean/120/' + str(num) + '_' + str(num_subplots) + '_120.npy', sub_data)
    else:
        np.save('./split_merge_class_clean/150/' + str(num) + '_' + str(num_subplots) + '_150.npy', sub_data)
```
