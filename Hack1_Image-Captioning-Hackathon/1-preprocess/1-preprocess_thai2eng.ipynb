```python
import pandas as pd
import json
from tqdm import tqdm
import ast
```

```python
tqdm.pandas()
```

```python
f = open('../../../dataset/coco/cocojson.json')

coco_json = json.load(f)
```

```python
df = pd.DataFrame({"keys": coco_json.keys(), "values": coco_json.values()})

df
```

```python
df.rename(columns={"keys": "image_id", "values": "thai"}, inplace=True)

df
```

```python
df["thai_1"] = df["thai"].progress_apply(lambda x: x[0])
df["thai_2"] = df["thai"].progress_apply(lambda x: x[1])
df["thai_3"] = df["thai"].progress_apply(lambda x: x[2])
```

```python
df.drop(columns=["thai"], inplace=True)

df
```

```python
train_eng = json.load(
    open('../../../dataset/coco/captions_train2017.json'))["annotations"]

train_eng_df = pd.DataFrame(train_eng)

train_eng_df.drop(columns=["id"], inplace=True)

train_eng_df['image_id'] = train_eng_df['image_id'].apply(
    lambda x: f"train2017/{x:012d}")

train_eng_df
```

```python
val_eng = json.load(
    open('../../../dataset/coco/captions_val2017.json'))["annotations"]

val_eng_df = pd.DataFrame(val_eng)

val_eng_df.drop(columns=["id"], inplace=True)

val_eng_df['image_id'] = val_eng_df['image_id'].apply(
    lambda x: f"val2017/{x:012d}")

val_eng_df
```

```python
combined_eng_df = pd.concat([train_eng_df, val_eng_df])
```

```python
caption_dict = dict(
    zip(combined_eng_df['image_id'], combined_eng_df['caption']))

df['eng'] = df['image_id'].map(caption_dict)

df
```
